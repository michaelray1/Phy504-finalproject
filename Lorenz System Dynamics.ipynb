# Lorenz System Dynamics

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Here I will present the results of solving the Lorenz system differential equations with various initial conditions and various parameter values.

The Lorenz system is a system of differential equations that are known to have chaotic solutions for certain values of parameters.

The system of equations is shown below with parameters $\sigma$, $\rho$, and $\beta$.

<center>$\frac{dx}{dt} = \sigma (y-x)$</center>

<center>$\frac{dy}{dt} = x (\rho - z) - y$</center>

<center>$\frac{dz}{dt} = xy-\beta z$</center>

The C++ code that comes alongside this Jupyter notebook contains an integrator to integrate this system of equations and solve for x(t), y(t), and z(t). The program then saves the data to a csv file, from which we can grab the data and plot it. Let's do this for one file. The name of the file may seem peculiar, but all it says is that $\rho = 28$, $\sigma = 10$, $\beta = 8/3$, $x_0 = 1$, $y_0 = 0$, and $z_0 = 0$.

```python
data100 = np.genfromtxt('motion_28_10_8over3_100.txt', delimiter=',')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20,7))
ax[0].plot(data100[:, 1], data100[:, 2])
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].plot(data100[:, 2], data100[:, 3])
ax[1].set_xlabel('y')
ax[1].set_ylabel('z')

ax[2].plot(data100[:, 1], data100[:, 3])
ax[2].set_xlabel('x')
ax[2].set_ylabel('z')

plt.suptitle('$x_0 = 1.0$, $y_0 = 0.0$, $z_0 = 0.0$')

fig2 = plt.figure(figsize=(20, 15))
ax2 = fig2.add_subplot(1,1,1, projection='3d')
ax2.plot(data100[:,1], data100[:,2], data100[:,3])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')
ax2.set_title('$x_0 = 1.0$, $y_0 = 0.0$, $z_0 = 0.0$')
```

We see above the "butterfly" characteristic behavior of the Lorenz system. When plotting x vs. z, we see that the solution resembles a butterfly.

When $\rho < 1$, it is known that the origin is a global attractor for the Lorenz system, so no matter where we start the particle at, it will be attracted to the origin. Let's see that in action. Below we will set up a particle with $\rho$ = 0.25 with a few different sets of initial conditions.

```python
data_rho25 = np.genfromtxt('motion_0p25_10_8over3_100.txt', delimiter=',')
datarho2 = np.genfromtxt("motion_0p25_10_8over3_0p500.txt", delimiter=',')
datarho3 = np.genfromtxt("motion_0p25_10_8over3_010.txt", delimiter=',')
```

```python
fig2 = plt.figure(figsize=(20, 15))
ax2 = fig2.add_subplot(1,3,1, projection='3d')
ax2.plot(data_rho25[:,1], data_rho25[:,2], data_rho25[:,3])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')
ax2.set_title('$x_0 = 1.0$, $y_0 = 0.0$, $z_0 = 0.0$')

ax2 = fig2.add_subplot(1,3,2, projection='3d')
ax2.plot(datarho2[:,1], datarho2[:,2], datarho2[:,3])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')
ax2.set_title('$x_0 = 0.5$, $y_0 = 0.0$, $z_0 = 0.0$')

ax2 = fig2.add_subplot(1,3,3, projection='3d')
ax2.plot(datarho3[:,1], datarho3[:,2], datarho3[:,3])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')
ax2.set_title('$x_0 = 0.0$, $y_0 = 1.0$, $z_0 = 0.0$')
```

We can see that no matter where we start the particle, it gets attracted to the origin, as it should.
